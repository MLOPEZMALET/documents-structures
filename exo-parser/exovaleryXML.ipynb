```python
xmlfile = "valery_ame-et-danse_1921.xml"
```

```python
tree = etree.parse(xmlfile)
```

```python
root = tree.getroot()
```

```python
print(etree.tostring(root, pretty_print=True))
```

```python
TEI_NAMESPACE = "http://www.tei-c.org/ns/1.0"
TEI = "{%s}" % TEI_NAMESPACE
```

```python
for element in root.iter(TEI+"edition"):
        print(element.tag, element.text, element.attrib)
```

```python
for element in root.iter(TEI+"licence"):
        print(element.tag, element.text, element.attrib)
```

```python
speakers = []
for element in root.iter(TEI+"label"):
        speakers.append(element.text)
```

```python
speakers
```

```python
collections.Counter(speakers).most_common(1)
```

```python
#autre solution
d = dict()
for element in root.iter(TEI + "label"):
    d[element.text] = d.get(element.text,0)+1 
max_speaker = max(d,key=d.get) 
#ou
sorted(d.items(),key=lambda x: x[1], reverse=True)
```

```python
for element in root.iter(TEI+"editionstmt"):
    new_elt = etree.SubElement(root,"respstmt")
    name = etree.SubElement(new_elt, "name")
    name.text("Marie-Chantal")
    resp = etree.SubElement(new_elt, "resp") 
    resp.text("par exemple")
```

```python
for element in root.iter(TEI+"signed"):
    for child in element:
        print(child.text.upper())
```
